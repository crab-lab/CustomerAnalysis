```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
import pandas as pd
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup
import random
```

```python
data= pd.read_csv("ContentCSV.csv")
```

```python
cols = ['_custom_selector','_h1','_h2','_h3', '_h4','_h5', '_h6','_p', '_td','_article']
data.fillna({'_custom_selector':'.','_h1':'.', '_h2':'.','_h3':'.', '_h4':'.','_h5':'.', '_h6':'.','_p':'.', '_td':'.','_article':'.'}, inplace=True)
```

```python
data[cols] = data[cols].replace('.*Not found..*', '.', regex=True)
```

```python
word = widgets.Text(placeholder='Enter keyword to search')
button_send = widgets.Button(
                description='Search for Keyword',
                tooltip='Search',
                style={'description_width': 'initial'}
            )
```

```python
URLarray = {""}
search1 = pd.DataFrame({""})
```

```python
def search_results():
    resultName = pd.unique(search1[['CompetitorName']].values.ravel())
    resultURL = pd.unique(search1[['URL']].replace('https://www.','').values.ravel())
    URLarray = np.array([resultURL], dtype = object)
    Result_array = np.array([resultName,resultURL], dtype=object)
    result = pd.DataFrame(Result_array)
    result = result.transpose()
    result.columns=['CompetitorName','URL']
output = widgets.Output()
def on_button_clicked(word):
    #search1 = data.loc[data['_p'].str.contains(word, case=False)]
    global search1 
    global word1
    search1 = data[data.apply(lambda row: row.astype(str).str.contains(word, case=False).any(), axis=1)]
    word1 = word
    search_results()
    with output:
        clear_output()
        print(search1)
button_send.on_click(on_button_clicked(word.value))
vbox_result = widgets.VBox([button_send, output])
```

```python
drop_down_value = ""
drop_down = widgets.Dropdown(options= ('No URL Selected'))
def drop_down_clicked(event):
    drop_down = widgets.Dropdown(options= URLarray)
    drop_down_value = drop_down.value
drop_down.observe(drop_down_clicked)
```

```python
query = "site:"+drop_down_value+" "+word1
```

```python
from serpapi import GoogleSearch
import os

params = {
    "api_key": "3bbd3b57017af5a80cedcb87fdf610e367c2c21d8a6d0b0f6d84f705a0897230", # serpapi API key
    "engine": "google",              # search engine
    "q": query,             # search query
    "hl": "en",                       # language of the search
    "start": "0",
    "num": "450",
}

search = GoogleSearch(params)      # where data extraction happens
results = search.get_dict()        # JSON -> Python dictionary
```

```python
for result1 in results['organic_results']:
    title = result1['title']
    link = result1['link']
    displayed_link = result1['displayed_link']
    snippet = result1['snippet']
    print(f"{title}\n{link}\n{displayed_link}\n{snippet}\n")
    print('---------------')
```
